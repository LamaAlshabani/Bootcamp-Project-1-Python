```python
import random

dict1 = {

}

board = []

Round= 0


def Menu(Round):
    while True:  
        print("Enter 1 to play")
        print("Enter 2 to display the scoreboard")
        print("Enter 3 to Exit")
        inp=int(input())
        if inp == 1 :
            Round+=1
            start(Round)
        elif inp == 2 :
            if len(dict1)!=0 :
                print(dict1)
            else :
                print("You must play first \n")
                
        elif inp == 3 :
            print("We hope you enjoyed playing , goodbye")
            break
        else :
            print("Invalid entry, try again")
        
def create_board():
        for i in range(3):
            row = []
            for j in range(3):
                row.append('-')
            board.append(row)

def get_random_first_player():
        return random.randint(0, 1)
def inputUser():
    row, col = list( map(int, input("Enter row and column numbers to fix spot: ").split()))
    print()
    if not ( (row >= 1 and row <= 3) and (col>=1 and  col <= 3)) :
        Menu(Round)
    return row, col

def fix_spot(row, col, player):
    
    if board[row][col]== "-" :
        board[row][col] = player
    else :
        print ("cheating! we will pass your turn to other player")

def is_player_win(player):
        win = None

        n = len(board)

        # checking rows
        for i in range(n):
            win = True
            for j in range(n):
                if board[i][j] != player:
                    win = False
                    break
            if win:
                return win

        # checking columns
        for i in range(n):
            win = True
            for j in range(n):
                if board[j][i] != player:
                    win = False
                    break
            if win:
                return win

        # checking diagonals
        win = True
        for i in range(n):
            if board[i][i] != player:
                win = False
                break
        if win:
            return win

        win = True
        for i in range(n):
            if board[i][n - 1 - i] != player:
                win = False
                break
        if win:
            return win
        return False

        for row in board:
            for item in row:
                if item == '-':
                    return False
        return True

def is_board_filled():
        for row in board:
            for item in row:
                if item == '-':
                    return False
        return True


swap_player_turn = lambda player : 'X' if player == 'O' else 'O'

def show_board():
        for row in board:
            for item in row:
                print(item, end=" ")
            print()

def start(Round):
        board.clear()
        create_board()

        player = 'X' if get_random_first_player() == 1 else 'O'
        while True:
            print("-------------------------------")
            print(f"Player {player} turn")

            show_board()

            # taking user input
#             row, col = list(
#                 map(int, input("Enter row and column numbers to fix spot: ").split()))
#             print()
#             if not ( (row >= 1 and row <= 3) and (col>=1 and  col <= 3)) :
#                 Menu(Round)
            row, col = inputUser()
            # fixing the spot
            fix_spot(row -1 , col -1 , player)

            # checking whether current player is won or not
            if is_player_win(player):
                print(f"Player {player} wins the game!")
                dict1[f"Round{Round}"] = f"{player} wins"
                break

            # checking whether the game is draw or not
            if is_board_filled():
                print("Match Draw!")
                Menu(Round)

            # swapping the turn
            player = swap_player_turn(player)

        # showing the final view of board
        print()
        show_board()

       

#starting the game
Menu(Round)

```
